```python
#PASSO3:
tabela.info() # - Me oferece informações presentes na tabela
#FLOAT - NUMERO COM CASA DECIMAL
#OBJECT - COLUNAS DE TEXTO

# valores vázios - excluir as linhas que tem valores vázios
tabela = tabela.dropna()

display(tabela.info())

#PASSO4:
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True))

#Em caso de formatação
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))

#PASSO5:
import plotly.express as px



#criando o gráfico
for coluna in tabela.columns:
grafico = px.histogram(tabela, x= coluna, color="cancelou") # Dá pra colar "duracao_contrato"



#exibe o grafico
grafico.show()

# todos is clientes que ligaram mais de 4x pro call center, cancelaram
# criar um processo interno
filtro = tabela["ligacoes_callcenter"]<=4
tabela = tabela[filtro]

filtro = tabela["dias_ataso"]<=20
tabela = tabela[filtro]

filtro = tabela["duracoes_contrato"]!="Monthly"
tabela = tabela[filtro]


display(tabela["cancelou"].value_counts(normalize=True))
```
